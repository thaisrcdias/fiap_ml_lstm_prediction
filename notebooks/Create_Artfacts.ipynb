```python
import pickle
from sklearn.pipeline import Pipeline
import yfinance as yf
import yaml
from models.utils.transformers import *

import warnings
warnings.filterwarnings("ignore")
```

```python
config = yaml.safe_load(open('config/config.yaml', 'rb'))
```

```python
features = config.get('features')
path_model = config.get('path_model')
path_scaler = config.get('path_scaler')
window_size = config.get('treino')[0].get('window_size')
model = pickle.load(open(path_model, 'rb'))
scaler = pickle.load(open(path_scaler, 'rb'))
```

```python
# Coleta dos dados
symbol = 'BTC-USD'
start_date = (datetime.now() - timedelta(days=70)).strftime('%Y-%m-%d')
end_date = datetime.now().strftime('%Y-%m-%d')

X = yf.download(symbol, start=start_date, end=end_date)
```

```python
create_dataframe = CreateDataFrame(features)
normalize = Normalize(scaler)
feature_engineering = FeatureEngineering(window_size)
model = Model(model, scaler)
```

```python
pipe = Pipeline([
        ('create_dataframe', create_dataframe), 
        ('normalize', normalize),
        ('feature_engineering',feature_engineering),
        ('model', model)])
```

```python
pipe
```

```python
score = pipe.predict(X)
```

```python
score
```

```python
with open('models/pipeline_model_LSTM.pkl', 'wb') as f:
    pickle.dump(pipe, f)
```
