```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pickle
```

```python
import yaml
```

# 2. Coleta dos dados

```python
symbol = 'BTC-USD'
start_date = '2018-01-01'
end_date = '2025-05-01'
df = yf.download(symbol, start=start_date, end=end_date)
df = df[['Close']].dropna()
```

# 3. Exploração dos dados (EDA)

```python
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['Close'], label='Preço de Fechamento')
plt.title('Preço de Fechamento do Bitcoin (BTC-USD)')
plt.xlabel('Ano')
plt.ylabel('Preço em USD')
plt.legend()
plt.show()

# Estatísticas rápidas
display(df.describe())
```

# 4. Normalização

```python
scaler = MinMaxScaler()
scaled = scaler.fit(df[['Close']])
data_scaled = scaled.transform(df[['Close']])
pickle.dump(scaled, open('../models/scaled.pkl', 'wb'))
```

# 5. Preparação dos dados para o LSTM

```python
def create_sequences(dataset, window_size):
    X, y = [], []
    for i in range(window_size, len(dataset)):
        X.append(dataset[i-window_size:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

window_size = 60
X, y = create_sequences(data_scaled, window_size)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

## 5.1 Separando treino e validação (80/20)

```python
split = int(0.8 * X.shape[0])
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]
```

# 6. Construção do Modelo LSTM

```python
def train_model(window_size, input_size, neuronios, dropout, loss_function, epochs, batch_size):
    '''
    Essa função é destinada ao treinamento da LSTM. Abaixo, a descrição dos parâmetros:
    - window_size: Janela anterior às variáveis atuais, utilizada para criação da entrada do modelo. 
    - input_size: Tamanho da entrada. Esse valor refere-se a quantidade de colunas que vão servir de entrada ao modelo.
    - neuronios: Quantidade de neuronios disponíveis na rede.
    - Dropout: Percentual de neoronios desligados na rede. Isso evita overfitting.
    - loss_function: métrica de validação, para otimização dos parâmetros.
    - epochs: Quantidade de épocas.
    - batch_size: Quantidade de dados utilizado em cada etapa do treinamento.
    - return_sequences=True: necessário pois temos 2 camadas LSTM
    - Dense(1): saída de um valor, o preço previsto
    '''
    
    model = Sequential([
        LSTM(neuronios, return_sequences=True, input_shape=(window_size, input_size)),
        Dropout(dropout),
        LSTM(neuronios),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss=loss_function)

    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val))

    # Visualizando a curva de aprendizado (Loss)
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Treinamento')
    plt.plot(history.history['val_loss'], label='Validação')
    plt.legend()
    plt.title('Curva de Loss do Modelo')
    plt.xlabel('Épocas')
    plt.ylabel('Loss')
    plt.show()

    return model


window_size=window_size
input_size=1
neuronios=50
dropout=0.2
loss_function='mean_absolute_error'
epochs=150
batch_size=32

model = train_model(
                    window_size=window_size, 
                    input_size=input_size, 
                    neuronios=neuronios, 
                    dropout=dropout, 
                    loss_function=loss_function,
                    epochs=epochs,
                    batch_size=batch_size
                  )
```

```python
pickle.dump(model, open('../models/modelo_LSTM.pkl', 'wb'))
```

```python
config = {
            'features':['Close']
            'path_scaler':'models/scaled.pkl',
            'path_model':'models/modelo_LSTM.pkl',
            'treino':[{'window_size':window_size, 
            'input_size':input_size, 
            'neuronios':input_size, 
            'dropout':dropout, 
            'loss_function':loss_function,
            'epochs':epochs,
            'batch_size':batch_size}]
}
```

# 7. Avaliação das previsões

```python
y_pred = model.predict(X_val)
y_pred_inv = scaler.inverse_transform(y_pred)
y_val_inv = scaler.inverse_transform(y_val.reshape(-1,1))

mae = mean_absolute_error(y_val_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_val_inv, y_pred_inv))
mape = np.mean(np.abs((y_val_inv - y_pred_inv) / y_val_inv)) * 100

print(f'MAE: {mae:.2f} | RMSE: {rmse:.2f} | MAPE: {mape:.2f}%')
```

```python
config.update({'metricas_validacao':[{'mae':round(mae, 2), 'rmse':round(float(rmse),2), 'mape':round(float(mape), 2)}]})
```

```python
with open('../config/config.yaml', 'w') as f:
    yaml.dump(config, f)
```

# 8. Visualizando previsão vs real

```python
plt.figure(figsize=(15,6))
plt.plot(y_val_inv, label='Real')
plt.plot(y_pred_inv, label='Previsto')
plt.title('Previsão x Real (Conjunto de Validação)')
plt.legend()
plt.show()
```

# 9. Teste (dados recentes)

```python
# Coleta dos dados
symbol = 'BTC-USD'
start_date = '2020-01-01'
end_date = '2025-07-17'
df_test = yf.download(symbol, start=start_date, end=end_date)
```

```python
df_test.reset_index(inplace=True)

df_test.columns =['Date'	, 'Close'	,'High'	,'Low'	,'Open'	,'Volume']
```

```python
data_scaled_test = scaled.transform(df_test[['Close']])

window_size = 60
X_test, y_test = create_sequences(data_scaled_test, window_size)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
# 9. Avaliação das previsões
y_pred_test = model.predict(X_test)
y_pred_inv_test = scaler.inverse_transform(y_pred_test)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
df_test_pred = pd.DataFrame()
df_test_pred['Date'] = df_test['Date'][60:].tolist()
df_test_pred['Close'] = df_test['Close'][60:].tolist()
df_test_pred['Close_pred'] = y_pred_inv_test
```

```python
df_test_pred = df_test_pred[df_test_pred['Date']>='2025-06-01']
```

```python
mae = mean_absolute_error(df_test_pred['Close'], df_test_pred['Close_pred'])
rmse = np.sqrt(mean_squared_error(df_test_pred['Close'], df_test_pred['Close_pred']))
mape = np.mean(np.abs((df_test_pred['Close'] - df_test_pred['Close_pred']) / df_test_pred['Close'])) * 100

print(f'MAE: {mae:.2f} | RMSE: {rmse:.2f} | MAPE: {mape:.2f}%')
```

```python
import matplotlib.pyplot as plt
# Plotando o preço de fechamento
plt.figure(figsize=(15,6))
plt.plot(df_test_pred['Date'], df_test_pred['Close'])
plt.plot(df_test_pred['Date'], df_test_pred['Close_pred'])

plt.title('Preço de Fechamento do Bitcoin (BTC-USD)')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento (USD)')
plt.ylim(20000, 140000)
plt.show()
```
